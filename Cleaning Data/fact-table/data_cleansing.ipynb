```python
# Fix Bulan dan Tahun Wisuda

df_wisuda = pd.read_excel("Wisuda.xlsx")

def check_fourth_and_fifth_chars(value):
    str_value = str(value)
    return str_value[0] == "1" and str_value[3:5] == "11"

df_wisuda_2011 = df_wisuda[df_wisuda['NIM'].apply(check_fourth_and_fifth_chars)]

df_wisuda_2011.drop(df_wisuda_2011.columns[[2, 3, 4, 5, 6]], axis=1, inplace=True)

wisuda_2011 = df_wisuda_2011.to_dict()

wisuda_nim = wisuda_2011['NIM']
wisuda_nama = wisuda_2011['Nama']
wisuda_tgl = wisuda_2011['Seremoni Wisuda']

final_data_tgl = {}
for i in wisuda_nama.keys():
    
    tgl = wisuda_tgl[i].lower().replace('wisuda','').lstrip().split(' ')
    
    if 'april' in tgl[0]:
        tgl[0] = '04'
    elif 'october' in tgl[0]:
        tgl[0] = '10'
    elif 'july' in tgl[0]:
        tgl[0] = '07'
    
    result = ''.join([tgl[0],'-',tgl[-1]])
    final_data_tgl[wisuda_nama[i]] = result


final_data_nim = {}
for i in wisuda_nama.keys():
    nim = str(wisuda_nim[i])
    final_data_nim[wisuda_nama[i]] = nim


for i in range(len(df_2018.index)):
    if df_2018.at[i, 'Nama'] in final_data_tgl:
        df_2018.at[i, 'Bulan dan Tahun Wisuda'] = str(final_data_tgl[df_2018.at[i, 'Nama']])

for i in range(len(df_2018.index)):
    if df_2018.at[i, 'Nama'] in final_data_nim:
        df_2018.at[i, 'NIM'] = str(final_data_nim[df_2018.at[i, 'Nama']])
```
